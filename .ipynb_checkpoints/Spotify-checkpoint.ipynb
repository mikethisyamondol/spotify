```python
feature_list = []
for i in results["tracks"]["items"]:
    features = sp.audio_features(tracks=i["id"])
    feature_list.append(features)
```

```python
attributes = []

for index in range(len(feature_list)):
    temp_list = []
    for feature_dict in feature_list[index]:
#             print(j)
            for keys in feature_dict:
                temp_list.append(feature_dict[keys])
    attributes.append(temp_list)

attributes = pd.DataFrame(attributes, columns=["danceability", "energy", "key", "loudness", "mode", "speechiness", "acousticness",
                                  "instrumentalness", "liveness", "valence", "tempo", "type", "id", "uri", "track_href",
                                  "analysis_url", "duration_ms", "time_signature"])
```

```python
attributes.head()
```

```python
name_lookup = []

for tracks in results["tracks"]["items"]:
    temp_dict = {"id" : tracks["id"], "name" : tracks["name"]}
    name_lookup.append(temp_dict)

name_lookup = pd.DataFrame(name_lookup)
```

```python
pop_lookup = []

for tracks in results["tracks"]["items"]:
    temp_dict = {"id" : tracks["id"], "popularity" : tracks["popularity"]}
    pop_lookup.append(temp_dict)

pop_lookup = pd.DataFrame(pop_lookup)
```

```python
album_lookup = []

for tracks in results["tracks"]["items"]:
    temp_dict = {"id" : tracks["id"], "album" : tracks["album"]["name"]}
    album_lookup.append(temp_dict)

album_lookup = pd.DataFrame(album_lookup)
album_lookup = album_lookup[["id", "album"]]
```

```python
attributes = attributes.merge(name_lookup, how = "inner", on = "id")
attributes = attributes.merge(album_lookup, how = "inner", on="id")
```

```python
attributes.set_index("name", inplace = True)
```

```python
attributes.head()
```

```python
attributes.drop(["type", "uri", "track_href", "analysis_url", "time_signature"], axis = 1, inplace = True)
```

```python
attributes.head()
```

```python
X = attributes.drop(["id", "album"], axis=1)
```

```python
ss = StandardScaler()
X_scaled = ss.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=5)
model = kmeans.fit(X_scaled)
```

```python
# model.cluster_centers_
```

```python
attributes['predicted'] = model.labels_
```

```python
attributes.sort_values("predicted", ascending=False)
```
